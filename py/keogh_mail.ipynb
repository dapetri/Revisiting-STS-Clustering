```python
samples_2000 = pd.read_csv('../data/scatter_samples_2000.csv',usecols=[1,2]).to_numpy()
samples_20 = pd.read_csv('../data/scatter_samples_20.csv',usecols=[1,2]).to_numpy()
```

### Running KMeans on both sampling sets two times and visualising clusters
First on data set with 2000 points.

```python
km_1 = KMeans(n_clusters=2).fit(samples_2000)
km_2 = KMeans(n_clusters=2).fit(samples_2000)

centroids_1 = km_1.cluster_centers_
centroids_2 = km_2.cluster_centers_

labels_1 = km_1.labels_
labels_2 = km_2.labels_

group_1_0 = np.zeros((np.count_nonzero(labels_1==0),2))
group_1_1 = np.zeros((np.count_nonzero(labels_1==1),2))

group_2_0 = np.zeros((np.count_nonzero(labels_2==0),2))
group_2_1 = np.zeros((np.count_nonzero(labels_2==1),2))

c_1 = 0
c_2 = 0
for i in range(len(labels_1)):
    if labels_1[i] == 0:
        group_1_0[c_1] = samples_2000[i]
        c_1 += 1
    else:
        group_1_1[c_2] = samples_2000[i]
        c_2 += 1

c_1 = 0
c_2 = 0
for i in range(len(labels_2)):
    if labels_2[i] == 0:
        group_2_0[c_1] = samples_2000[i]
        c_1 += 1
    else:
        group_2_1[c_2] = samples_2000[i]
        c_2 += 1

groups = ('Cluster 1', 'Cluster 2', 'Centroids')

plt.figure()
plt.scatter(group_1_0[:,0],group_1_0[:,1],s=np.pi*2,c='blue',label=groups)
plt.scatter(group_1_1[:,0],group_1_1[:,1],s=np.pi*2,c='red',label=groups)
plt.scatter(centroids_1[:,0],centroids_1[:,1],s=100,marker='x',c='black',label=groups)
plt.title('First run of KMeans')

plt.figure()
plt.scatter(group_2_0[:,0],group_2_0[:,1],s=np.pi*2,c='blue')
plt.scatter(group_2_1[:,0],group_2_1[:,1],s=np.pi*2,c='red')
plt.scatter(centroids_2[:,0],centroids_2[:,1],s=100,marker='x',c='black')
plt.title('Second run of KMeans')
```
